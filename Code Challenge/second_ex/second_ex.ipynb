# Definition 
In this exercise we will be testing your multithreading skills. 
Suppose that you want to simulate a fruit farm: 
* Three farmers are collecting fruits from a single tree to a single dirty fruit basket. 
* In parallel, three other farmers are getting the fruits from the dirty fruit basket, cleaning them, and pushing 
them into the single cleaned fruit basket. 
* All the farmers are managing the fruit individually 
* The tree has 50 fruits (and only one farmer at one time can pick fruit from the tree) 
* Time to collect fruits from the trees into the basket: random(3,6) seconds 
* Time to clean the fruits into the cleaned fruit basket: random(2,4) seconds 
* The simulation ends when all the fruits from the tree are collected and cleaned. 
* The number of fruits in the tree and in the baskets must be logged every second.

# Expected deliverable 
Your code should be executable with the following call “yourScript.py”, with a similar output log 
to the one below: 
* 2020-12-01 19:02:00 Tree (50 fruits) - dirty basket ( 0 ) - Clean Basket ( 0 ) - farmer1 
(0) - farmer2 (0) - cleaner1(0) - cleaner2 (0)
* 2020-12-01 19:03:00 Tree (45 fruits) - dirty basket ( 3 ) - Clean Basket ( 1 ) - farmer1 
(1) - farmer2 (0) - cleaner1(0) - cleaner2 (0) 
* 2020-12-01 19:10:00 Tree (0 fruits) - dirty basket ( 0 ) - Clean Basket ( 50 ) - farmer1 
(0) - farmer2 (0) - cleaner1(0) - cleaner2 (0)

```python
import random
import time
import threading
from datetime import datetime

class Fruit:
    def __init__(self, name):
        self.name = name

class Tree:
    def __init__(self, num_fruits):
        self.fruits = [Fruit(f"fruit{i}") for i in range(1, num_fruits + 1)]
        self.lock = threading.Lock()

    def collect_fruit(self):
        with self.lock:
            if self.fruits:
                return self.fruits.pop()

class Basket:
    def __init__(self):
        self.fruits = []
        self.lock = threading.Lock()

    def add_fruit(self, fruit):
        with self.lock:
            self.fruits.append(fruit)

    def remove_fruit(self):
        with self.lock:
            if self.fruits:
                return self.fruits.pop()

def farmer_work(farmer_name, tree, dirty_basket):
    while True:
        fruit = tree.collect_fruit()
        if fruit:
            time.sleep(random.uniform(3, 6))
            dirty_basket.add_fruit(fruit)

def cleaner_work(cleaner_name, dirty_basket, clean_basket):
    while True:
        fruit = dirty_basket.remove_fruit()
        if fruit:
            time.sleep(random.uniform(2, 4))
            clean_basket.add_fruit(fruit)

def print_status(tree, dirty_basket, clean_basket):
    while True:
        now = datetime.now().strftime("%Y-%m-%d %H:%M:%S")
        tree_size = len(tree.fruits)
        dirty_size = len(dirty_basket.fruits)
        clean_size = len(clean_basket.fruits)

        print(f"{now} Tree ({tree_size} fruits) - dirty basket ({dirty_size}) - Clean Basket ({clean_size}) "
              f"- farmer1 (0) - farmer2 (0) - farmer3 (0) - cleaner1(0) - cleaner2 (0) - cleaner3 (0)")
        time.sleep(1)

if __name__ == "__main__":
    tree = Tree(50)
    dirty_basket = Basket()
    clean_basket = Basket()

    farmer_threads = []
    cleaner_threads = []

    for i in range(1, 4):
        farmer_thread = threading.Thread(target=farmer_work, args=(f"farmer{i}", tree, dirty_basket))
        farmer_threads.append(farmer_thread)

        cleaner_thread = threading.Thread(target=cleaner_work, args=(f"cleaner{i}", dirty_basket, clean_basket))
        cleaner_threads.append(cleaner_thread)

    status_thread = threading.Thread(target=print_status, args=(tree, dirty_basket, clean_basket))

    for thread in farmer_threads + cleaner_threads + [status_thread]:
        thread.start()

    for thread in farmer_threads + cleaner_threads + [status_thread]:
        thread.join()
```
